```python
import numpy as np
from scipy.stats import spearmanr
for col in df.columns:
    pearsonCorr = np.corrcoef(df['crimeperpop'],df[col])[0][1]
    #regCorr = df['crime count'].corr(df[col])
    corr,pVal =  spearmanr(df['crimeperpop'],df[col])
    print('col: {0},corr: {1},pVal: {2},pearson: {3}'.format(col,corr,pVal,pearsonCorr))
    print('')
```
